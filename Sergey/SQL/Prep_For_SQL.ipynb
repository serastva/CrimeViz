```python
import os 
import pandas as pd
from datetime import datetime
```

# #Crime_DF  SQL cleanup


```python
crime_df = pd.read_csv("../Output/4_crime_data_lat_long_tract.csv")

crime_df
```

```python
crime_df[['Street_No', 'Latitude','Longitude','Census_Tract']] = crime_df[['Street_No', 'Latitude','Longitude','Census_Tract']].fillna(0) 
crime_df.dtypes
```

```python
crime_df
```

```python
crime_df['Street_No'] = crime_df['Street_No'].replace('.0', '', regex=True)
crime_df
```

```python
#Convert Datatypes
crime_df['Incident'] = crime_df['Incident'].astype('string')
crime_df['Date'] = pd.to_datetime(crime_df['Date'])
crime_df['NIBRS_Class'] = crime_df['NIBRS_Class'].astype('string')
crime_df['Offense_Type'] = crime_df['Offense_Type'].astype('string')
crime_df['Offense_Grouped'] = crime_df['Offense_Grouped'].astype('string')
crime_df['Beat'] = crime_df['Beat'].astype('string')
crime_df['Premise'] = crime_df['Premise'].astype('string')
crime_df['Street_No'] = crime_df['Street_No'].astype('string')
crime_df['Street_Name'] = crime_df['Street_Name'].astype('string')
crime_df['Street_Type'] = crime_df['Street_Type'].astype('string')

crime_df['City'] = crime_df['City'].astype('string')
crime_df['State'] = crime_df['State'].astype('string')
crime_df['Full_Address'] = crime_df['Full_Address'].astype('string')
crime_df['Street_Address'] = crime_df['Street_Address'].astype('string')
crime_df['Census_Tract'] = crime_df['Census_Tract'].astype('string')

crime_df.dtypes
```

# #Create a DF called Criminal_Activity for SQL 

```python
criminal_activity_df = crime_df[["Incident","Date","Hour","Offense_Type","Offense_Grouped","Beat","Premise","Street_No","Street_Name","Street_Type","Full_Address" ]]


criminal_activity_df["ID"] = criminal_activity_df.index + 1
criminal_activity_df
```

```python
criminal_activity_df['Full_Address'] = criminal_activity_df['Full_Address'].replace(', Houston, TX', '', regex=True)
criminal_activity_df
```

```python
criminal_activity_df.dtypes
```

```python
criminal_activity_df.to_csv('criminal_activity_df.csv', index=False)
```
