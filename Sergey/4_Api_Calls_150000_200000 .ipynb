```python
import pandas as pd
```

```python
crime_cleaned_df = pd.read_csv('Output/3_crime_data_lat_long.csv')
crime_cleaned_df
```

```python
address_list = crime_cleaned_df['Street_Address'].to_list()
city_list = crime_cleaned_df['City']
```

```python
address_list
```

```python
city_list
```

```python
import requests
import csv

tract = []
counter = 0

for street, city in zip(address_list[150000:200000], city_list[150000:200000]):
    try:
        base_url = f'https://geocoding.geo.census.gov/geocoder/geographies/address?street={street}&city={city}&state=TX&benchmark=Public_AR_Census2020&vintage=Census2020_Census2020&layers=10&format=json'
        response = requests.get(base_url)
        tract_json = response.json()
        tract.append(tract_json['result']['addressMatches'][0]['geographies']['Census Blocks'][0]['TRACT'])
        print(f"{counter} {street} Tract: {tract_json['result']['addressMatches'][0]['geographies']['Census Blocks'][0]['TRACT']}")
        print('-------------')                                        
    except:
        print(f'{counter} {street}: failed')
        print('-------------')  
        tract.append('NaN')
    
    counter+=1


```

```python
tract_df = pd.DataFrame(tract, columns=['Census_Tract'])
```

```python
tract_df.to_csv('Output/Tract_150000_200000.csv', index=False,)
```
