```python
#Import dependencies 
import pandas as pd
import os
import datetime as dt
import numpy as np
```

## Load data and do initial clean up 

```python
# read in merged file 
crime_df = pd.read_csv("Output/1_crime_data_merged.csv")
crime_df
```

```python
#Return the shape of the array
print(crime_df.shape)
```

```python
#Find out what data types are assigned to each column
crime_df.dtypes
```

```python
crime_df['Incident'] = crime_df['Incident'].astype('stro')
```

```python
#Replace NaN in Street type with a blank space

crime_df['Street_Type'] = crime_df['Street_Type'].replace(np.nan, '', regex=True)

crime_df = crime_df.astype(str)

crime_df
```

```python
#Remove the suffix column
del crime_df['Suffix']
crime_df.head()
```

## Group offenses by type  

```python
#Check all crimes in the list
crime_types = crime_df['Offense_Type'].to_list()
crime_types
```

```python
#Check all unique crimes in the list
crime_df.Offense_Type.unique()
```

```python
#Create a crime list of only violent crimes discard misdemeanors or nonviolent crimes
# CRIMES NOT INCLUDED: 
#'False pretenses, swindle', 'Credit card, ATM fraud','Impersonation','Welfare fraud','Wire fraud',
#'Identify theft', 'Disorderly conduct','Driving under the influence', 'Drunkenness','Liquor law violations',
#'All other offenses','Betting/wagering', 'Promoting gambling','Peeping tom','Gambling equipment violations',
#'Curfew, loitering, vagrancy violations','Runaway', 'Family offenses, no violence','Animal Cruelty','Trespass of real property', 
#'Pornographs, obscene material','Justifiable homicide,'Bribery', 'Extortion, Blackmail','Embezzlement', 'Counterfeiting, forgery',
#'Bad checks','Hacking/Computer Invasion',

crime_list = [  'Theft from motor vehicle',
                'Theft of motor vehicle parts or accessory', 
                'All other larceny',
                'Motor vehicle theft', 
                'Pocket-picking',
                'Purse-snatching',
                'Theft from building',
                'Stolen property offenses', 
                'From coin-operated machine or device', 
                'Shoplifting', 
                'Aggravated Assault',
                'Simple assault', 
                'Intimidation', 
                'Forcible rape', 
                'Forcible sodomy', 
                'Statutory rape',
                'Forcible fondling', 
                'Human Trafficking/Commercial Sex Act', 
                'Sexual assault with an object',
                'Incest', 
                'Robbery', 
                'Burglary, Breaking and Entering',
                'Murder, non-negligent',
                'Negligent manslaughter',
                'Human Trafficking/Involuntary Servitude', 
                'Kidnapping, abduction',
                'Drug, narcotic violations', 
                'Drug equipment violations', 
                'Destruction, damage, vandalism',
                'Arson', 
                'Prostitution', 
                'Purchasing prostitution',
                'Assisting or promoting prostitution', 
                'Weapon law violations',
            ]
```

```python
#Create a new dataframe with only crimes which are in the crime_list
violent_crime_df = crime_df[crime_df['Offense_Type'].isin(crime_list)]
violent_crime_df.head()
```

```python
print(violent_crime_df.shape)
```

```python
violent_crime_df['Offense_Type'].value_counts()
```

```python
#Group Crimes by type

#Theft
theft = ['Theft from motor vehicle','Theft of motor vehicle parts or accessory','All other larceny','Motor vehicle theft', 
'Pocket-picking','Purse-snatching','Theft from building','Stolen property offenses','From coin-operated machine or device', 
'Shoplifting']

#Assault
assault = ['Aggravated Assault', 'Simple assault','Intimidation']

#Sex Crime
sex_crime = ['Forcible rape','Forcible sodomy','Statutory rape','Forcible fondling','Human Trafficking/Commercial Sex Act', 
'Sexual assault with an object','Incest']

#Burglary
burglary_robbery = ['Burglary, Breaking and Entering', 'Robbery',]

#Murder
murder = ['Murder, non-negligent','Negligent manslaughter',]

#Abduction
abduction = ['Kidnapping, abduction','Human Trafficking/Involuntary Servitude']

#Narcotics
narcotics = ['Drug, narcotic violations','Drug equipment violations']

#Vandalism
vandalism = ['Destruction, damage, vandalism']

#Arson
arson = ['Arson']

#Prostitution
prostitution = ['Prostitution','Purchasing prostitution','Assisting or promoting prostitution']

#Weapons
weapons = ['Weapon law violations']
```

```python
#Create a new column for grouped offenses
violent_crime_df['Offense_Grouped']=''
violent_crime_df.head()
```

```python
#Output grouped offenses to a new column called offense_Grouped 
violent_crime_df['Offense_Grouped'][violent_crime_df['Offense_Type'].isin(theft)] = 'Theft'
violent_crime_df['Offense_Grouped'][violent_crime_df['Offense_Type'].isin(assault)] = 'Assault'
violent_crime_df['Offense_Grouped'][violent_crime_df['Offense_Type'].isin(sex_crime)] = 'Sex Crime'
violent_crime_df['Offense_Grouped'][violent_crime_df['Offense_Type'].isin(burglary_robbery)] = 'Burglary/Robbery'
violent_crime_df['Offense_Grouped'][violent_crime_df['Offense_Type'].isin(murder)] = 'Murder/Manslaughter'
violent_crime_df['Offense_Grouped'][violent_crime_df['Offense_Type'].isin(abduction)] = 'Abduction'
violent_crime_df['Offense_Grouped'][violent_crime_df['Offense_Type'].isin(narcotics)] = 'Narcotics'
violent_crime_df['Offense_Grouped'][violent_crime_df['Offense_Type'].isin(vandalism)] = 'Vandalism'
violent_crime_df['Offense_Grouped'][violent_crime_df['Offense_Type'].isin(arson)] = 'Arson'
violent_crime_df['Offense_Grouped'][violent_crime_df['Offense_Type'].isin(prostitution)] = 'Prositution'
violent_crime_df['Offense_Grouped'][violent_crime_df['Offense_Type'].isin(weapons)] = 'Weapons'
violent_crime_df
```

## Clean Zip Codes


```python
violent_crime_df.Zip_Code.unique()
```

```python
#Manually clean Zip codes with a - 
violent_crime_df['Zip_Code'] = violent_crime_df['Zip_Code'].replace('-0000', '', regex=True)
violent_crime_df['Zip_Code'] = violent_crime_df['Zip_Code'].replace('-3942', '', regex=True)
violent_crime_df['Zip_Code'] = violent_crime_df['Zip_Code'].replace('-4103', '', regex=True)
violent_crime_df['Zip_Code'] = violent_crime_df['Zip_Code'].replace('-4006', '', regex=True)
violent_crime_df['Zip_Code'] = violent_crime_df['Zip_Code'].replace('-2920', '', regex=True)
violent_crime_df['Zip_Code'] = violent_crime_df['Zip_Code'].replace('-2921', '', regex=True)
violent_crime_df['Zip_Code'] = violent_crime_df['Zip_Code'].replace('-5505', '', regex=True)
violent_crime_df['Zip_Code'] = violent_crime_df['Zip_Code'].replace('-1469', '', regex=True)
violent_crime_df.Zip_Code.unique()
```

```python
#Remove zipcodes that were not in houston and nan
zip_list = ['78582',
            'nan',
            '78572',
            '78504',
            '78501',
            '78557',
            '78596',
            '78343',
            '78414',
            '78418',
            '78414',
            '77979',
            '78941',
            '70601',
            '77701',
            '77566',
            '77486',
            '77448',
            '77356',
            '37013',
            '77662',
            '75961',
            '77879',
            '76033',
            '75050',
            '75234',
            '75075',
            '77879',
            '77437',
            '77378']
```

```python
#Remove Zip codes on the zip_list from the violent_crime_df
violent_crime_df = violent_crime_df[~violent_crime_df.Zip_Code.isin(zip_list)]
```

# Fix Addresses and add three new columns one for Texas, Street address and one for full address

```python
violent_crime_df['City'].value_counts()
```

```python
#Convert all cities to Houston
violent_crime_df['City']=violent_crime_df['City'].map('Houston'.format)
violent_crime_df['City'].value_counts()
```

```python
violent_crime_df['City'].value_counts()
```

```python
#Add a column called state and add TX to it for geolocation
violent_crime_df['State'] = 'TX'
violent_crime_df.head()
```

```python
#Splits Street_No that are given in a range
violent_crime_df['Street_No'] = violent_crime_df['Street_No'].apply(lambda x: x.split('-')[0])
violent_crime_df
```

```python
#Create a full address using the street_no, Street_name, Street_type, City, and state
violent_crime_df['Full_Address'] = violent_crime_df['Street_No'] + ' ' + violent_crime_df['Street_Name'] + ' ' + violent_crime_df['Street_Type'] + ', ' + violent_crime_df['City'] + ', ' + violent_crime_df['State']

violent_crime_df.head()
```

```python
address_list = violent_crime_df['Full_Address'].to_list()
print(len(address_list))
```

```python
#Create a full address using the street_no, Street_name, Street_type, also remove spaces and replace with + 

violent_crime_df['Street_Address'] = violent_crime_df['Street_No'] + '+' + violent_crime_df['Street_Name'] + '+' + violent_crime_df['Street_Type'] 
violent_crime_df['Street_Address'] = violent_crime_df['Street_Address'].replace(' ', '+', regex=True)
violent_crime_df.head()
```

```python
violent_crime_df.dtypes
```

```python
#convert objects 

#datetime
violent_crime_df['Date'] = pd.to_datetime(violent_crime_df['Date'])
#incident to int
violent_crime_df['Incident'] = violent_crime_df['Incident'].astype(int)
#Offense count to Int
violent_crime_df['Offense_Count'] = violent_crime_df['Offense_Count'].astype(int)
```
